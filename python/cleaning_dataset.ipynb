#Import Dataset

```python
import pandas as pd

foo = pd.read_csv('adult_dataset.csv')
foo
```

#Explore Data

```python
columns = {
    'work_class': foo['workclass'].sort_values().unique(),
    'education': foo['education'].sort_values().unique(),
    'education_num': foo['education-num'].sort_values().unique(),
    'marital_status': foo['marital-status'].sort_values().unique(),
    'occupation': foo['occupation'].sort_values().unique(),
    'relationship': foo['relationship'].sort_values().unique(),
    'race': foo['race'].sort_values().unique(),
    'sex': foo['sex'].sort_values().unique(),
    'native_country': foo['native-country'].sort_values().unique(),
    'income': foo['income'].sort_values().unique()
}

for key, values in columns.items():
  print(key + ':', values)
```

```python
foo.isnull().sum()
```

#Convert Invalid Values to None

```python
bar = foo.copy()
```

```python
bar.loc[bar['workclass'] == '?', 'workclass'] = None
bar.loc[bar['occupation'] == '?', 'occupation'] = None
bar.loc[bar['native-country'] == '?', 'native-country'] = None

bar.loc[bar['workclass'].isnull(), 'workclass'] = None
bar.loc[bar['occupation'].isnull(), 'occupation'] = None
bar.loc[bar['native-country'].isnull(), 'native-country'] = None

bar.info()
bar.isnull().sum()
```

```python
bar[(bar['workclass'].notnull()) & (bar['occupation'].isnull())]
```

```python
bar[bar['native-country'].isnull()]
```

#Clean Values in "Income" Column
| Orignal Value | Converted Value |
| ------------- | --------------- |
| <50K          | <=50K           |
| <=50K.        | <=50K           |
| >50K          | >50K            |
| >50K.         | >50K            |

```python
bar.loc[bar['income'] == '<50K', 'income'] = '<=50K'
bar.loc[bar['income'] == '<=50K.', 'income'] = '<=50K'
bar.loc[bar['income'] == '>50K.', 'income'] = '>50K'

bar['income'].unique()
```

#Drop Null Values

```python
foobar = bar.copy()

foobar = foobar[foobar['occupation'].notnull()]
foobar = foobar[foobar['native-country'].notnull()]

foobar
```

#Drop Columns

```python
baz = foobar.copy()

baz = baz.drop(['fnlwgt', 'education'], axis=1)
baz
```

#Categorical Variable Encoding

```python
qux = pd.get_dummies(baz, drop_first=True)
qux.head(3)
```

#Write to file

```python
qux.to_csv('clean_adult_dataset.csv')
```
